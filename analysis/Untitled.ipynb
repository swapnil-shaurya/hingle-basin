```python
import numpy as np
import pandas as pd
from os.path import join, dirname, isfile, basename
from os import getcwd, listdir
```

```python
oil = pd.read_csv('ooip.csv')
```

```python
eastings = [ 
    4700,
    4700,
    57000,
    58000,
    59000,
    60000,
    61000,
    62000,
    63000,
    64000,
]
northings = [
    400,
    800,
    10000,
    8000,
    12000,
    10000,
    8000,
    12000,
    10000,
    8000,
]
```

```python
step = oil['northing'][1]-oil['northing'][0]
```

```python
well_length = 5280
c = []
for e, n in zip(eastings, northings):
    c.append(np.power(10, oil.loc[(oil['northing']-step <= n) & (n <= oil['northing']+step ) & (oil['easting'] >= e) & 
                  (oil['easting'] <= e+well_length)]['OOIP'].values).sum())
```

```python
s = 0
for x in c:
    s+=x
```

```python
data_address = join(dirname(getcwd()), "mapping")
all_data = pd.read_csv(join(data_address, "all_data.csv"))
```

```python
p_a = []
o_a = []
for e, n in zip(eastings, northings):
    rows = all_data.loc[(oil['northing']-step <= n) & (n <= oil['northing']+step ) & (oil['easting'] >= e) & 
                  (oil['easting'] <= e+well_length)]
    p_a.append(rows['porosity'].mean())
    o_a.append(rows['oil saturation'].mean() - 0.0005)
```

```python
for o, p, e, n in zip(o_a, p_a, eastings, northings):
    print("Well originating at: {}E, {}N, porosity average: {}, oil saturation average:{}".format(
    e,n,p,n))
```
